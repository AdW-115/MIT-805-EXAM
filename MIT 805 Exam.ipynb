```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('South Africa-forecasted-life-expectancy-line.csv')
```

```python
df
```

```python
df_male = df[df['Sex'] == 'Males']
df_female = df[df['Sex'] == 'Females']
```

```python
df_female
```

```python
df_male
```

```python
x = list(df_male['Year'])
y_male = list(df_male['Age'])
y_female = list(df_female['Age'])
```

```python
plt.plot(x, y_male, label = 'Male life expetancy')
plt.plot(x, y_female, label = 'Female life expetancy')
plt.xlabel("Year")
plt.ylabel("Age")
plt.title('South African life expectancy and forecast')
plt.legend()
plt.savefig('life_expectancy.png', bbox_inches='tight')
plt.show()
```
